```python
import pandas as pd

# Load the files
monthly_df = pd.read_csv(r"C:\Users\Admiin\OneDrive\Documents\air_quality_health_monthly.csv")
cities_df = pd.read_csv(r"C:\Users\Admiin\OneDrive\Documents\cities.csv")

# Merge them on the 'city' column
# This adds State, Zone, Population, and Industrialization to every monthly row
combined_df = pd.merge(monthly_df, cities_df, on='city', how='left', suffixes=('', '_metadata'))

# Fix the date (Crucial Step)
combined_df['date'] = pd.to_datetime(combined_df[['year', 'month']].assign(day=1))

print(combined_df.head())
combined_df.to_csv('master_air_quality_dataset.csv', index=False)
```

```python
# Drop columns that end with '_metadata'
cols_to_drop = [c for c in combined_df.columns if c.endswith('_metadata')]
combined_df = combined_df.drop(columns=cols_to_drop)

# Double check the result
print(combined_df.columns)

# Save this as your final Clean Master file
combined_df.to_csv('master_india_air_quality.csv', index=False)
```
